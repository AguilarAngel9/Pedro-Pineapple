```python
import os
import pandas as pd
import matplotlib.pyplot as plt

# Get the current directory
current_directory = os.getcwd()

# List all files in the current directory
files = [file for file in os.listdir(current_directory) if file.endswith('.csv')]

# Create an empty dictionary to store dataframes
dataframes = {}

# Iterate through CSV files and create dataframes
for file in files:
    df = pd.read_csv(file)
    dataframes[file] = df

# Extract year information from filenames for legend
legend_labels = {}
for filename in dataframes.keys():
    parts = filename.split('_')
    year_range = parts[1].split('.')[0]  # Extracting the year range from the filename
    legend_labels[filename] = f"SPY {year_range[:4]} to {year_range[4:]}"

# Plotting
plt.figure(figsize=(10, 6))

for filename, df in dataframes.items():
    plt.plot(df['Close'], label=legend_labels[filename])

plt.xlabel('Time (days)')
plt.ylabel('Value (Close price)')
plt.title('Close Values Comparison')
plt.xlim(0,1200)
plt.legend()
plt.grid()
plt.show()
```
